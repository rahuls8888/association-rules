In the below code cells, we can observe that we made a trail by changing the values of min_support to 0.05, 0.08, 0.1,0.15,0.2, but there is no change in the 0.05,0.08,0.1 and there is no change in the 0.15,0.2. So, we are going to consider the min_support as 0.05.

```python
frequent_itemsets=apriori(df1,min_support=0.05,use_colnames=True)
frequent_itemsets.info()
```

```python
frequent_itemsets_1=apriori(df1,min_support=0.08,use_colnames=True)
frequent_itemsets_1.info()
```

```python
frequent_itemsets_2=apriori(df1,min_support=0.1,use_colnames=True)
frequent_itemsets_2.info()
```

```python
frequent_itemsets_3=apriori(df1,min_support=0.15,use_colnames=True)
frequent_itemsets_3.info()
```

Here you can observe that the min_support is 0.05 and the graphs which represents the values

```python
rules=association_rules(frequent_itemsets,metric="lift")
rules
```

```python
rules.sort_values(by='lift',ascending=False)
```

```python
rules[['support','confidence','lift']].hist(figsize=(15,15))
plt.show()
```

```python
rules[['support']].boxplot(figsize=(15,5))
plt.show()
```

```python
rules[['confidence']].boxplot(figsize=(15,5))
plt.show()
```

```python
rules[['lift']].boxplot(figsize=(15,5))
plt.show()
```

```python
plt.scatter(rules['support'],rules['confidence'])
plt.show()
```

```python
plt.scatter(rules['confidence'],rules['lift'])
plt.show()
```

```python
plt.scatter(rules['lift'],rules['support'])
plt.show()
```

Here we can observe that the min_support is 0.15, and also the graphs which is going to represent the values.
